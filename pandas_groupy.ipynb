```python
grouped_df = df.groupby(["Region","SalesPerson"])["UnitPrice"].agg(sum).reset_index()
grouped_df
```

```python
grouped_df["sum"] = grouped_df.groupby("Region")["UnitPrice"].transform("sum")
grouped_df
```

```python
grouped_df["perc_cont"] = (grouped_df["UnitPrice"]/grouped_df["sum"])*100
grouped_df
```

```python
df
```

```python
grouped_df = df.groupby(["Region","SalesPerson"]).agg(
    total_sales = ("UnitPrice","sum")
).reset_index()
grouped_df
```

```python
region_sales_df = grouped_df.groupby("Region").agg(
    region_total_sales=("total_sales","sum")
).reset_index()

region_sales_df
```

```python
grouped_df=grouped_df.merge(region_sales_df,on="Region",how="left")
grouped_df
```

```python
grouped_df["perc_cont"] = (grouped_df["total_sales"]/grouped_df["region_total_sales"])*100
grouped_df
```

```python
df1 = grouped_df[0:4]
df1
```
